```python
def split_lyrics(lyrics):
    all_song_elements = []
    for i in lyrics:
        artist = i[0]
        song_name = i[1]
        all_lyrics = i[2]
        split_lyrics = all_lyrics.split('\n\n')

        for i in split_lyrics:
            try:
                if i[:2] != '\n':
                    verse = i.split('\n')
                    if verse == '':
                        song_element = ''
                        only_lyrics = ''
                    else:
                        if verse[0][0] == '[':
                            song_element = verse[0]
                            only_lyrics = verse[1:]
                        else:
                            song_element = ''
                            only_lyrics = verse
            except:
                song_element = ''
                only_lyrics = ''
            all_song_elements.append([artist,song_name,song_element,only_lyrics])

    return all_song_elements
```

```python
def lyrics_to_df(lyrics):
    columns = ['song_owner','song_title','song_element','verse']
    df = pd.DataFrame(lyrics, columns = columns)
    df['verse'] = df['verse'].apply(', '.join)
    return df
```
